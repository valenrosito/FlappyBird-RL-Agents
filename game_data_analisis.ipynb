```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv('game_data/states_20251118_195738.csv')
```

### Descripción de las columnas del estado
----
- `episode`: número de episodio.  
- `step`: paso dentro del episodio.  
- `reward`: recompensa obtenida en ese paso.  
- `total_reward`: recompensa acumulada del episodio hasta ese paso.  
- `player_y`: posición vertical (Y) del jugador.  
- `player_vel`: velocidad vertical del jugador.  
- `next_pipe_dist_to_player`: distancia horizontal al próximo tubo.  
- `next_pipe_top_y`: coordenada Y de la parte superior del próximo tubo.  
- `next_pipe_bottom_y`: coordenada Y de la parte inferior del próximo tubo.  
- `next_next_pipe_dist_to_player`: distancia al segundo tubo.  
- `next_next_pipe_top_y`: coordenada Y de la parte superior del segundo tubo.  
- `next_next_pipe_bottom_y`: coordenada Y de la parte inferior del segundo tubo.

### Información Recolectada
---
- `reward` = -5 -> indica que se termino el juego
- `reward` = 1 -> representa que se supero una tuberia
- `player_vel` < 0 -> Indica que estamos cayendo
- `player_y` = 30 -> indica el suelo
- `player_y` = 400 -> indica el techo

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.hist(datos['player_y'], bins=50, color='blue', alpha=0.7)
plt.title('Distribución de player_y')
plt.xlabel('player_y')
plt.ylabel('Frecuencia')    
plt.subplot(2, 2, 2)
plt.hist(datos['player_vel'], bins=50, color='green', alpha=0.7)
plt.title('Distribución de player_vel')
plt.xlabel('player_vel')
plt.ylabel('Frecuencia')    
plt.subplot(2, 2, 3)
plt.hist(datos['next_pipe_dist_to_player'], bins=50, color='red', alpha=0.7)
plt.title('Distribución de next_pipe_dist_to_player')
plt.xlabel('next_pipe_dist_to_player')
plt.ylabel('Frecuencia')    
plt.subplot(2, 2, 4)
plt.hist(datos['next_next_pipe_dist_to_player'], bins=50, color='purple', alpha=0.7)
plt.title('Distribución de next_next_pipe_dist_to_player')
plt.xlabel('next_next_pipe_dist_to_player')
plt.ylabel('Frecuencia')    
plt.tight_layout()
plt.show()
```
